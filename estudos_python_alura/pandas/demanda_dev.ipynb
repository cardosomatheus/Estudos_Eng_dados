```python
import pandas as pd 
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv', sep=';')
dados.head(100)
```

```python
# adicionar a valor mensal e anual do imovel

dados[['Valor','Condominio','IPTU']] = dados[['Valor','Condominio','IPTU']].fillna(0)


dados['Valor_por_mes'] = dados['Valor']+dados['Condominio']
dados['Valor_por_ano'] = (dados['Valor_por_mes']*12) + dados['IPTU']
dados
```

```python
# descricao do imovel  e se o imovel possui Suite
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados
```

```python
# Se o imovel possui Suite
dados['Possui_suite'] = dados['Suites'].apply(lambda x: 'Sim' if x>0 else 'Não')
dados
```
